```python
# Exibir a contagem
# Contar o número de acidentes por tipo
# saída leito carroçavel é a pista e acostamento
contagem_acidentes_por_tipo = dados_acidente['tipo_acidente'].value_counts()
print(contagem_acidentes_por_tipo)
```

```python
# Criar o gráfico de barras Contar os Tipos de Acidentes
plt.figure(figsize=(18, 6))
sns.barplot(x=contagem_acidentes_por_tipo.index, y=contagem_acidentes_por_tipo.values, palette='viridis')
plt.title('Tipos de Acidentes')
plt.xlabel('Tipo de Acidente')
plt.ylabel('Número de Acidentes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Contar o número de acidentes por estado
# SP concentra 46 Milhões de pessoas e MG 21,3 Milhões - com isso, é possível afirmar que minas tem as estradas mais propensa a acidentes no Brasil
acidentes_por_estado = dados_acidente['uf'].value_counts()
print(acidentes_por_estado)
```

```python
# Criar o gráfico de barras
plt.figure(figsize=(12, 6))  # Ajustar o tamanho da figura
acidentes_por_estado.plot(kind='bar', color='skyblue')
plt.title('Número de Acidentes por Estado')
plt.xlabel('Estado')
plt.ylabel('Número de Acidentes')
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x para melhor legibilidade
plt.grid(axis='y')  # Adicionar uma grade apenas no eixo y
plt.tight_layout()  # Ajustar o layout para evitar sobreposição
plt.show()  # Exibir o gráfico
```

```python
# Criar uma coluna apenas com o mês e o ano
dados_acidente['month_year'] = dados_acidente['data_inversa'].dt.to_period('M')
```

```python
# Converter 'month_year' para string para evitar possíveis problemas de compatibilidade no gráfico
dados_acidente['month_year'] = dados_acidente['month_year'].astype(str)
```

```python
# Agrupar o número de acidentes apenas por mês
df_grouped = dados_acidente.groupby('month_year').size().reset_index(name='id')

# Evolução de total de todos as ocorrências por mês
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_grouped, x='month_year', y='id')
plt.xticks(rotation=90)
plt.title('Evolução de Acidentes por Mês')
plt.ylabel('Quantidade de Acidentes')  # Rotular o eixo Y
plt.xlabel('Mês/Ano')  # Rotular o eixo X

# Ajustar o eixo Y para começar em 0
plt.ylim(bottom=0)

# Adicionar rótulos com os valores no gráfico
for index, row in df_grouped.iterrows():
    plt.text(row['month_year'], row['id'], 
             row['id'], 
             color='black', 
             ha='center', 
             va='bottom')

plt.show()
```

```python
# Contar o número de acidentes por  classificacao_acidente

distribuicao_classificacao_acidente = dados_acidente['classificacao_acidente'].value_counts()
print(distribuicao_classificacao_acidente)
```

```python
# Contar a distribuição de classificacao_acidente
distribuicao_classificacao_acidente = dados_acidente['classificacao_acidente'].value_counts()

# Criar o gráfico de pizza
plt.figure(figsize=(10, 8))  # Ajustar o tamanho da figura
plt.pie(distribuicao_classificacao_acidente, labels=distribuicao_classificacao_acidente.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribuição de Tipos de Acidentes')
plt.axis('equal')  # Para garantir que o gráfico de pizza seja um círculo
plt.show()  # Exibir o gráfico
```
